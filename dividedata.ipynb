# Importing and division of data:

````python
import numpy as np
from IPython.display import display,HTML
from numpy import *

np.set_printoptions(suppress=True)

from IPython.display import Markdown, display
def printmd(string):
    display(Markdown(string))
````

#### Importing the data

The data was imported using the numpy function ```genfromtxt('file.csv')``` which reads and converts ```.csv``` files into a single ```numpy array``` which will be useful for dividing and saving purposes.

````python
### Importing some useful packages
#This ensures that plots appear in the sheet and not in a new window
%matplotlib inline 

#Pandas is an alternative numerical processing library like numpy: they have differences
#We will use Pandas only to provide a preview of the data once we've read it in
import pandas as pd


data = np.genfromtxt('online_shoppers_intention_selection.csv',delimiter=",")

#Convert the numpy array to a pandas DataFrame
#and display the first 10 rows as html
display(pd.DataFrame(data).head(10))
````

#### Randomizing the data 

It is important to randomize ones data set before dividing the data into Training, Cross Validtion and Testing data sets. Thus the numpy function ```numpy.random.shuffle(arr)``` was used to shuffle the data by rows while leaving the contents of each row unshifted.

````python
np.random.shuffle(data)
display(pd.DataFrame(data).head(10))
````

#### Dividing the data

The origional dataset was divided into 3 datasets which are namely, ```Training``` consisting of 60% of the origional data, ```Cross Validation``` consisting of 20% and ```Testing``` consisting of 20% of remaining origional data. 

````python
#sets the ratio of training data: validation data: testing data(currently 80:10:10)

perc1= int(0.6*len(data))
perc2= int(0.8*len(data))

train = data[:perc1]
val = data[perc1:perc2]
test = data[perc2:]
````

````python
print("\nThere is a total of "+(str)(len(data))+ " records in this data set")

#Convert the numpy array to a pandas DataFrame
#and display the first 10 rows as html for each data set

print("\nTraining data: "+(str)(len(train)))
display(pd.DataFrame(train).head(10))
print("\nCV data: "+(str)(len(val)))
display(pd.DataFrame(val).head(10))
print("\nTesting data: "+(str)(len(test)))
display(pd.DataFrame(test).head(10))
````

#### Saving the datasets

The divided datasets were then stored into their respective ```.dat``` files, which was made possible through the use of the function ```savetxt('file.dat')```

````python
#writes the arrays into their respective files
savetxt('data/train.dat', train)
savetxt('data/cv.dat', val)
savetxt('data/test.dat', test)
````
